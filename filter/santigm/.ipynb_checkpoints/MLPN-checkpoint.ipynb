```python
trn_df = pd.read_csv('filter/datosEntrenamiento.csv', header=None)
tst_df = pd.read_csv('filter/datosPrueba.csv', header=None)
tst_idx = trn_df.shape[0] # guardar donde empiezan los datos de prueba
full_df = pd.concat([trn_df, tst_df]) #juntar datos de entrenamiento y prueba
full_df = (full_df - full_df.min())/(full_df.max() - full_df.min()) # escalar en [0,1](MLPN)
full_df.head(10)
```

```python
def eval_model(y_hat, y):
    n_elem_sel = y_hat[y_hat==1].size
    true_pos = y[(y==1)&(y==y_hat)].size
    false_neg = y[(y==0)&(y!=y_hat)].size
    false_pos = y[(y==1)&(y!=y_hat)].size
    true_neg = y[(y==0)&(y==y_hat)].size
    
    precision = true_pos/n_elem_sel
    tasa_v_pos = true_pos/(true_pos+false_neg)
    tasa_f_pos = false_pos/(true_neg+false_pos)
    return precision, tasa_v_pos, tasa_f_pos
```
